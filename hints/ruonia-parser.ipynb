```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://5ka-sale.ru/cities/'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
cities = []
rows = soup.table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    cities.append([ele for ele in cols if ele])
pd.DataFrame(cities).to_csv('cities.csv', sep=';')   
```
